IMPORTING REQUIRED PACKAGES FOR THE MODEL BUILDING

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_percentage_error

from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

```python
def perfromance(lab, pred):
    a = mean_squared_error(lab, pred) 
    p =root_mean_squared_error (lab, pred) 
    r =mean_absolute_percentage_error (lab, pred)
    return pd.DataFrame({"mean_sqrt_error": [a], "root_mean_sqrt_error": [p], "mean_absolute_percentage_error": [r]})
```

```python
df1=pd.read_csv("power_consumption_cleaned.csv")
df1.info()
```

```python
val=df1.drop(columns=["Year","Month","peak_hours","power_consumption"],axis=1)
tar=df1["power_consumption"]
```

```python
#TRAINING AND TESTING DATA
# Split the data into training and testing sets
trdata, tedata, trlab, telab = train_test_split( val, tar, test_size= 0.20, random_state= 46 )
```

TRAINING THE MODEL

```python
model= LinearRegression()
# Fit the model to the training data
model.fit(trdata, trlab)
```

```python
#TRAININ DATA PREDICTION
train_pred= model.predict(trdata)
train_pred.shape
```

```python
#TESTING DATA PREDICTION
ts_pred= model.predict(tedata)
ts_pred.shape
```

```python
perfromance(trlab, train_pred)
```

```python
perfromance(telab, ts_pred)
```

```python
import pickle
# Save the model to disk
model
with open("linera_model.pkl","wb")as f:
    pickle.dump(model,f)
```

```python
with open("Linera_model.pkl","rb") as f:
    lin_model=pickle.load(f)
lin_model
```

Random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Create a Random Forest Regressor model
# Initialize the model
RF_model = RandomForestRegressor(random_state=46, n_jobs=-1)
RF_model.fit(trdata, trlab)
```

```python
tra_pred=RF_model.predict(trdata)
tes_pred=RF_model.predict(tedata)
```

GRADENT BOOSTING

```python
from sklearn.ensemble import GradientBoostingRegressor
# Create a Gradient Boosting Regressor model
# Initialize the model
GB_model = GradientBoostingRegressor(random_state=46)
GB_model
```

```python
#TRAINING THE MODEL
GB_model.fit(trdata, trlab)
```

```python
#TRAINNIN PREDICTION
tra_pred=GB_model.predict(trdata)
#TESTING PREDICTION
tes_pred=GB_model.predict(tedata)
```

```python
#performance of the model
perfromance(trlab, tra_pred)
print(perfromance(trlab, tra_pred))
perfromance(telab, tes_pred)
print(perfromance(telab, tes_pred))
```
